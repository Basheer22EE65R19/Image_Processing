```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def translate(image, dx, dy):
    rows, cols = image.shape
    translated_image = np.zeros((rows, cols), dtype=np.uint8)
    
    for y in range(rows):
        for x in range(cols):
            new_x = x + dx
            new_y = y + dy
            
            if 0 <= new_x < cols and 0 <= new_y < rows:
                translated_image[new_y, new_x] = image[y, x]
    
    return translated_image
```

```python
def rotate(image, angle):
    rows, cols = image.shape
    rotated_image = np.zeros((rows, cols), dtype=np.uint8)
    center_x = cols // 2
    center_y = rows // 2
    
    for y in range(rows):
        for x in range(cols):
            new_x = int((x - center_x) * np.cos(angle) - (y - center_y) * np.sin(angle) + center_x)
            new_y = int((x - center_x) * np.sin(angle) + (y - center_y) * np.cos(angle) + center_y)
            
            if 0 <= new_x < cols and 0 <= new_y < rows:
                rotated_image[new_y, new_x] = image[y, x]
    
    return rotated_image
```

```python
def scale(image, sx, sy):
    rows, cols = image.shape
    scaled_image = np.zeros((int(rows * sy), int(cols * sx)), dtype=np.uint8)
    
    for y in range(rows):
        for x in range(cols):
            new_x = int(x * sx)
            new_y = int(y * sy)
            scaled_image[new_y, new_x] = image[y, x]
    
    return scaled_image
```

```python
def shear(image, shear_x, shear_y):
    rows, cols = image.shape
    sheared_image = np.zeros((rows, cols), dtype=np.uint8)
    
    for y in range(rows):
        for x in range(cols):
            new_x = int(x + shear_x * y)
            new_y = int(y + shear_y * x)
            
            if 0 <= new_x < cols and 0 <= new_y < rows:
                sheared_image[new_y, new_x] = image[y, x]
    
    return sheared_image
```

```python
# Load an example image (replace with your image)
image_path = "Lena.jpg"
image = Image.open(image_path).convert("L")
input_image = np.array(image)

# Apply different transformations
translated = translate(input_image, 50, 50)
rotated = rotate(input_image, np.pi / 4)
scaled = scale(input_image, 1.5, 0.8)
sheared = shear(input_image, 0.2, 0)

# Plot original and transformed images
plt.figure(figsize=(12, 10))

plt.subplot(2, 3, 1)
plt.imshow(input_image, cmap='gray')
plt.title("Original Image")

plt.subplot(2, 3, 2)
plt.imshow(translated, cmap='gray')
plt.title("Translated")

plt.subplot(2, 3, 3)
plt.imshow(rotated, cmap='gray')
plt.title("Rotated")

plt.subplot(2, 3, 4)
plt.imshow(scaled, cmap='gray')
plt.title("Scaled")

plt.subplot(2, 3, 5)
plt.imshow(sheared, cmap='gray')
plt.title("Sheared")

plt.tight_layout()
plt.show()
```
